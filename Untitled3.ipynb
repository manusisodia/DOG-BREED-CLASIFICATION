```python
import os
import cv2
import numpy as np

# Define the path to your dataset
path = '/content/data/training_data'

# Lists to store images and labels
images = []
labels = []

# List all directories (categories) in the path
dir_list = os.listdir(path)

# Iterate over each directory
for i in dir_list:
    dir = os.path.join(path, i)  # Get the full path of the directory

    # List all files in the directory
    file_list = os.listdir(dir)

    # Iterate over each file in the directory
    for j in file_list:
        files = os.path.join(dir, j)  # Get the full path of the file

        # Read, resize, normalize, and append the image
        img = cv2.imread(files)
        img = cv2.resize(img, (64, 64))
        img = np.array(img, dtype=np.float32)
        img = img / 255.0  # Normalize the image

        images.append(img)  # Append the processed image to the list
        labels.append(i)  # Append the corresponding label (directory name)
```

```python
X = np.array(images)
y = np.array(labels)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.utils import shuffle
X_sh, y_sh = shuffle(X, y, random_state=42)
```

```python
model = Sequential()
model.add(Conv2D(filters=16, kernel_size=(3,3), activation='relu', input_shape=(64,64,3)))
model.add(MaxPooling2D())
model.add(Conv2D(filters=32, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(filters=128, kernel_size=(3,3), activation='relu'))
model.add(Flatten())
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=36, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics = ['accuracy'])
history = model.fit(X_sh, y_sh ,validation_split=0.2, batch_size=16, epochs=10)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['loss', 'val_loss'])
```

```python
import os
import cv2
import numpy as np

# Lists to store test images and labels
test_images = []
test_labels = []

# Path to the testing data
path = '/content/data/testing_data'

# List all directories (categories) in the path
dir_list = os.listdir(path)

# Iterate over each directory
for i in dir_list:
    dir = os.path.join(path, i)  # Get the full path of the directory

    # List all files in the directory
    file_list = os.listdir(dir)

    # Iterate over each file in the directory
    for j in file_list:
        files = os.path.join(dir, j)  # Get the full path of the file

        # Read, resize, normalize, and append the image
        img = cv2.imread(files)
        img = cv2.resize(img, (64, 64))
        img = np.array(img, dtype=np.float32)
        img = img / 255.0  # Normalize the image

        test_images.append(img)  # Append the processed image to the list
        test_labels.append(i)  # Append the corresponding label (directory name)
```

```python
X_test = np.array(test_images)
y_test = np.array(test_labels)
```

```python
preds = model.predict(X_test)
predicted_labels = le.inverse_transform(np.argmax(preds, axis=1))
```

```python
y_test = le.fit_transform(y_test)
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy}")
```
